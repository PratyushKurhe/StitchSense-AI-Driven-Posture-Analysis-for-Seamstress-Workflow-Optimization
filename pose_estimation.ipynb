```python
import imageio
import cv2
from PIL import Image, ImageDraw, ImageFont
import ultralytics
from ultralytics import YOLO
model=YOLO("yolov8n-pose.pt")    

cap=cv2.VideoCapture("D:\\ResoluteAI\\Pose_Estimation\\ProjectVid1.mp4")
fps=cap.get(cv2.CAP_PROP_FPS)

writer=imageio.get_writer("output_new.mp4", fps=fps)
frame_count=cap.get(cv2.CAP_PROP_FRAME_COUNT)
frame_number=0
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
success, image1=cap.read()


while success:
    results=model(image1)
    if results[0].keypoints.data[0][9][0]>=728 and results[0].keypoints.data[0][9][0]<=1102 and results[0].keypoints.data[0][9][1]>=755 and results[0].keypoints.data[0][9][1]<=1044 :
        image1[37: 157, 448:700]=255 #Only here first there is Y-axis then X-axis 
        cv2.imwrite("result.png",image1)
        image=Image.open("result.png")
        draw=ImageDraw.Draw(image)
        font=ImageFont.truetype("arial.ttf", 30)
        draw.text((448, 37),"Sewing", fill="black",font=font, align="left") 
        draw.text((448, 74),f"Frame no.: {frame_number}", fill="black",font=font, align="left") 
        image.save("texted.png")
        image1=cv2.imread("texted.png")
        writer.append_data(image1)
        frame_number+=1
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        success, image1=cap.read()
    elif results[0].keypoints.data[0][9][0]>=1201 and results[0].keypoints.data[0][9][0]<=1480 and results[0].keypoints.data[0][9][1]>=384 and results[0].keypoints.data[0][9][1]<=750 :
        image1[37: 157, 448:700]=255 
        cv2.imwrite("result.png",image1)
        image=Image.open("result.png")
        draw=ImageDraw.Draw(image)
        font=ImageFont.truetype("arial.ttf", 30)
        draw.text((448,37),"Placing", fill="black",font=font, align="left") 
        draw.text((448, 74),f"Frame no.: {frame_number}", fill="black",font=font, align="left") 
        image.save("texted.png")
        image1=cv2.imread("texted.png")
        writer.append_data(image1)
        frame_number+=1
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        success, image1=cap.read()
    elif results[0].keypoints.data[0][9][0]>=446 and results[0].keypoints.data[0][9][0]<=1480 and results[0].keypoints.data[0][9][1]>=30 and results[0].keypoints.data[0][9][1]<=1060 :
        image1[37: 157, 448:700]=255 
        cv2.imwrite("result.png",image1)
        image=Image.open("result.png")
        draw=ImageDraw.Draw(image)
        font=ImageFont.truetype("arial.ttf", 30)
        draw.text((448,37),"Material Handling", fill="black",font=font, align="left") 
        draw.text((448, 74),f"Frame no.: {frame_number}", fill="black",font=font, align="left") 
        image.save("texted.png")
        image1=cv2.imread("texted.png")
        writer.append_data(image1)
        frame_number+=1
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        success, image1=cap.read()
    else:
        print("Ok")

writer.close()



```
